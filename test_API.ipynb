```python
from llama_index import SimpleDirectoryReader

documents = SimpleDirectoryReader(
    input_dir="./data",
).load_data()
```

```python
documents
```

```python
from llama_index.node_parser import SimpleNodeParser

node_parser = SimpleNodeParser.from_defaults(chunk_size=5000)
base_nodes = node_parser.get_nodes_from_documents(documents)
```

```python
for i in range(len(base_nodes)):
    print(base_nodes[i].get_text())
```

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
chroma_client.delete_collection(name="test")
```

```python
collection = chroma_client.create_collection(name="test", metadata={"hnsw:space": "cosine"})
```

```python
base_nodes[0].metadata.get('file_name')#.get_text()
```

```python
collection.add(
    #embeddings=[embeddings_batch_response.data[0].embedding, embeddings_batch_response.data[1].embedding],
    documents=[base_nodes[0].get_text(), base_nodes[1].get_text(), base_nodes[2].get_text(), base_nodes[3].get_text()],
    metadatas=[{'source ': base_nodes[0].metadata.get('file_name')}, {'source ': base_nodes[1].metadata.get('file_name')},
                {'source ': base_nodes[2].metadata.get('file_name')}, {'source ': base_nodes[3].metadata.get('file_name')}],
    ids=['0', '1', '2', '3']
)
```

```python
dbresults = collection.query(
            query_texts="Quels sont les bénifices net de l'entreprise sur la période 2023 ?",
            n_results = 4
        )
```

```python
from mistralai.models.chat_completion import ChatMessage
from mistralai.client import MistralClient
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("API_KEY")

model = "mistral-tiny"

client = MistralClient(api_key=api_key)

messages = [
    ChatMessage(role="user", content=f"Quels sont les bénéfices net ? :{dbresults.get('documents')[0][0]} ?")
]

chat_response = client.chat(
    model=model,
    messages=messages,
    max_tokens=100
)
```

```python
from IPython.display import display, Markdown

display(Markdown(f"{chat_response.choices[0].message.content}"))
```

# API connection and first chat completion

```python
from dotenv import load_dotenv
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
import os

model = "mistral-tiny"

load_dotenv()

api_key = os.getenv("API_KEY")

client = MistralClient(api_key=api_key)

messages = [
    ChatMessage(role="user", content="Who is the creator of Tesla?")
]

chat_response = client.chat(
    model=model,
    messages=messages,
    max_tokens=30
)

chat_response
```
